# 4. Forensic Summary & Recommendations - Titanic

**Goal:** Synthesize findings into historical and modern forensic insights.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset('titanic')
```

## 1. Matrix Analysis

High-fidelity view of survival rates across groups.

```python
pivot = df.pivot_table(index='sex', columns='pclass', values='survived', aggfunc='mean')
sns.heatmap(pivot, annot=True, fmt='.2%', cmap='RdYlGn')
plt.title('Survival Rate Heatmap')
plt.show()
```

## 2. Forensic Summary

1. **Protocol Adherence**: 'Women and Children First' was the strongest predictor.
2. **Wealth Effect**: Class status disproportionately affected survival beyond protocol.
